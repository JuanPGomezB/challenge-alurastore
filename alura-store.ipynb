```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

"""
Descomentar las siguientes lineas si se quiere cargar los archivos desde la computadora local
"""

# url = "./base-de-datos-challenge1-latam/tienda_1.csv"
# url2 = "./base-de-datos-challenge1-latam/tienda_2.csv"
# url3 = "./base-de-datos-challenge1-latam/tienda_3.csv"
# url4 = "./base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

```python
tienda2.head(15)
```

# 1. Análisis de facturación



```python
ingresost1=tienda['Precio'].sum()
ingresost2=tienda2['Precio'].sum()
ingresost3=tienda3['Precio'].sum()
ingresost4=tienda4['Precio'].sum()
print(f'ingresos tienda 1: {ingresost1}')
print(f'ingresos tienda 2: {ingresost2}')
print(f'ingresos tienda 3: {ingresost3}')
print(f'ingresos tienda 4: {ingresost4}')
maximo=max(ingresost1,ingresost2,ingresost3,ingresost4)
print(f'La tienda con mayores ingresos es la tienda 1 con {maximo}')   
```

# 2. Ventas por categoría

```python
#Contar el número de productos vendidos por tienda por categoria
cantp_VendidosT1 = tienda['Categoría del Producto'].groupby(tienda['Categoría del Producto']).count().sort_values(ascending=False)
cantp_VendidosT2 = tienda['Categoría del Producto'].groupby(tienda2['Categoría del Producto']).count().sort_values(ascending=False)
cantp_VendidosT3 = tienda['Categoría del Producto'].groupby(tienda3['Categoría del Producto']).count().sort_values(ascending=False)
cantp_VendidosT4 = tienda['Categoría del Producto'].groupby(tienda4['Categoría del Producto']).count().sort_values(ascending=False)

print(f'Cantidad de productos vendidos por tienda 1: {cantp_VendidosT1}')
print(f'Cantidad de productos vendidos por tienda 2: {cantp_VendidosT2}')
print(f'Cantidad de productos vendidos por tienda 3: {cantp_VendidosT3}')
print(f'Cantidad de productos vendidos por tienda 4: {cantp_VendidosT4}')
```

# 3. Calificación promedio de la tienda


```python
cal_promT1 = tienda['Calificación'].mean()
cal_promT2 = tienda2['Calificación'].mean()
cal_promT3 = tienda3['Calificación'].mean()
cal_promT4 = tienda4['Calificación'].mean()

print(f'Calificación promedio de productos vendidos por tienda 1: {cal_promT1}')
print(f'Calificación promedio de productos vendidos por tienda 2: {cal_promT2}')
print(f'Calificación promedio de productos vendidos por tienda 3: {cal_promT3}')
print(f'Calificación promedio de productos vendidos por tienda 4: {cal_promT4}')
```

# 4. Productos más y menos vendidos

```python
Producto_mas_vendidoT1 = tienda['Producto'].groupby(tienda['Producto']).count().sort_values(ascending=False).head(1)
Producto_mas_vendidoT2 = tienda2['Producto'].groupby(tienda2['Producto']).count().sort_values(ascending=False).head(1)
Producto_mas_vendidoT3 = tienda3['Producto'].groupby(tienda3['Producto']).count().sort_values(ascending=False).head(1)
Producto_mas_vendidoT4 = tienda4['Producto'].groupby(tienda4['Producto']).count().sort_values(ascending=False).head(1)

Producto_menos_vendidoT1 = tienda['Producto'].groupby(tienda['Producto']).count().sort_values(ascending=True).head(1)
Producto_menos_vendidoT2 = tienda2['Producto'].groupby(tienda2['Producto']).count().sort_values(ascending=True).head(1)
Producto_menos_vendidoT3 = tienda3['Producto'].groupby(tienda3['Producto']).count().sort_values(ascending=True).head(1)
Producto_menos_vendidoT4 = tienda4['Producto'].groupby(tienda4['Producto']).count().sort_values(ascending=True).head(1)

print(f'Producto más vendido en tienda 1: {Producto_mas_vendidoT1}')
print(f'Producto más vendido en tienda 2: {Producto_mas_vendidoT2}')
print(f'Producto más vendido en tienda 3: {Producto_mas_vendidoT3}')
print(f'Producto más vendido en tienda 4: {Producto_mas_vendidoT4}')

print(f'Producto menos vendido en tienda 1: {Producto_menos_vendidoT1}')
print(f'Producto menos vendido en tienda 2: {Producto_menos_vendidoT2}')
print(f'Producto menos vendido en tienda 3: {Producto_menos_vendidoT3}')
print(f'Producto menos vendido en tienda 4: {Producto_menos_vendidoT4}')
```

```python
df1 = [Producto_mas_vendidoT1.values[0], Producto_mas_vendidoT2.values[0], Producto_mas_vendidoT3.values[0], Producto_mas_vendidoT4.values[0]]
df2 = [Producto_menos_vendidoT1.values[0], Producto_menos_vendidoT2.values[0], Producto_menos_vendidoT3.values[0], Producto_menos_vendidoT4.values[0]]
dfnames1 = [Producto_mas_vendidoT1.index[0], Producto_mas_vendidoT2.index[0], Producto_mas_vendidoT3.index[0], Producto_mas_vendidoT4.index[0]]
dfnames2 = [Producto_menos_vendidoT1.index[0], Producto_menos_vendidoT2.index[0], Producto_menos_vendidoT3.index[0], Producto_menos_vendidoT4.index[0]]
index = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
df = pd.DataFrame({'ProductoMasvendido': df1,
                   'ProductoMenosvendido': df2}, index=index)
ax = df.plot.bar(rot=0)
for i in range(len(df1)):
    ax.text(i-0.14, 15, str(dfnames1[i][:15])+".", ha='center', va='center', rotation='vertical')
    ax.text(i+0.14, 15, str(dfnames2[i][:15])+".", ha='center', va='center', rotation='vertical')
# print(Producto_mas_vendidoT1.index[0])
plt.show()
```
